# Show optimization result with camera angle movement

```python
import numpy as np
import matplotlib.pyplot as plt
import os

dirpath = '../outputs/invert_infloasion/cube_grid_angle'
angle_min, angle_max = -15, 10
num_cols = 3
num_rows = int((angle_max - angle_min) / num_cols) + 1
fig, ax = plt.subplots(num_rows, num_cols, figsize=(3*num_cols, 2*num_rows))
itr = 0
for angle in range(angle_min, angle_max+1):
    img_path = os.path.join(dirpath, 'angle{}'.format(angle))
    losses = np.loadtxt(os.path.join(img_path, 'losses.csv'), delimiter=',', encoding="utf-8")
    row, col = int(itr/num_cols), itr%num_cols
    ax[row, col].plot(np.arange(len(losses)), losses)
    itr += 1
    
```
